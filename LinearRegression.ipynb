```python
import numpy as np
import pandas as pd
import math
import operator
import matplotlib.pyplot as plt
# %matplotlib inline

data = pd.read_csv("data.csv")
data
x = data['x'].values
y = data['y'].values


mean_x = np.mean(x)
mean_y = np.mean(y)
x,y,mean_x,mean_y
m = len(x)

numer = 0 
denom = 0
for i in range(m):
  numer += (x[i] - mean_x)*(y[i]-mean_y)
  denom += (x[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)

print(f'm = {m} \nc = {c}')

max_x = np.max(x)+5
min_x = np.min(y)-5

X = np.linspace (min_x, max_x, 100)
Y = c + m * X

plt.plot(X, Y, color='#58b970', label='Regression Line')
plt.scatter(x, y, c='#ef5423', label='data points')

plt.xlabel('Independent X')
plt.ylabel('Dependent Y')
plt.legend()
plt.show()
```
